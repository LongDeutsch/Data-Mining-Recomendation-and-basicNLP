```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_arr = te.fit_transform(data_set, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(
                te_arr, columns=[str(c) for c in te.columns_])
sparse_df.head()
```

### Áp dụng apriori cho asociation rule mining

```python
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(data_set, sparse=False)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
freq_items = apriori(df, min_support=0.05, 
                     use_colnames=True)
freq_items
```

+ Do dữ liệu khá nhỏ gần 10000 rows so với unique values (khoảng 7000 unique values) nên độ hỗ trợ của các tập dữ liệu tương đối thấp
+ Do số lượng unique values khá cao với số lượng rows của dữ liệu, mà tần suất xuất hiện của các unique values khá thấp ở mỗi row nên support của các tập dữ liệu tương đối thấp

```python
# Kiểm tra thử support của một tập dữ liệu
k=0
count=0
for i in data_set:
    for j in i:
        if j=='bottled beer':
            count+=1
count/9834
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(freq_items, metric='lift',
                 min_threshold=1.01)
```

##### comments:
+ Mọi người mua nhiều nhất (whole milk) khi vào grocery
+ Xác suất khi mua whole milk trong khi trước đó đã mua yogurt là cao nhất (conf = 0.4). Khả năng mua whole milk + yogurt cao gấp 1.57 lần chỉ mua riêng lẻ whole milk hoặc yogurt (lift = 1.57)
+ Nhìn chung độ thuyết phục (conviction) của các tập dữ liệu còn khá thấp nên không tin tưởng để đánh giá sự phụ thuộc của các groceries
